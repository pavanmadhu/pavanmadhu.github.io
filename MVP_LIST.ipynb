```python

def parse_jsonp(jsonp_str):
    try:
        start = jsonp_str.index('(') + 1
        end = jsonp_str.rindex(')')
        return json.loads(jsonp_str[start:end])
    except ValueError as e:
        raise ValueError(f"Not a valid JSONP: {e}")

# Assuming jsonp_response is the byte string from your JSONP response
jsonp_response = r.content

# Convert to string
jsonp_str = jsonp_response.decode()

# Use the custom JSONP parser
data = parse_jsonp(jsonp_str)



# prompt: how to access a dataframe by column name

df['franchise'].unique()
df['points'] = 0

# prompt: how to get index of df matching value from one column
for ele in data["mvp"]:
  # print(ele["PlayerID"],ele["PlayerName"],ele["IndexValue"])
  index = df.index[df['hash'] == ele['PlayerID']]
  if(len(index)>0):
    # print(index[0])
    df.loc[index[0],'points'] = float(ele["IndexValue"])
    # print(df[index[0],'points'])
# Group the DataFrame by 'id' and sum the 'value' column

grouped_df = pd.DataFrame((df.groupby('franchise'))['points'].sum())

```
